###Data Preprocessing 

**Importing the libraries**

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

**Importing dataset**

```python
ds= pd.read_csv('Data.csv')
```

```python
ds
```

```python
ds.head()
```

**Handling the missing data**

```python
ds.isnull().sum()
```

```python
ds.Age=ds.Age.fillna(round(ds.Age.mean(),0))
ds
```

```python
ds.Salary=ds.Salary.fillna(round(ds.Salary.mean(),-3))
ds
```

**categorical data**

```python
le=LabelEncoder()
ds['Purchased']=le.fit_transform(ds['Purchased'])
```

```python
ds.Purchased=ds.Purchased.replace({"No":0,"Yes":1})
ds
```

**Creating a dummy variable**

```python
country=pd.get_dummies(ds.Country)
country
```

```python
ds1=pd.concat([ds,country],axis=1)
ds1=ds1.drop(columns=["Country"])
ds1
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
X=ds1.drop(columns=["Purchased"])
X
```

```python
y=ds1.Purchased
y
```

**Scaling Method using xtrain ytrain**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
scaler = StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```
